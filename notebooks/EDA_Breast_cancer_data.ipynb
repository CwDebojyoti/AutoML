# Importing the libraries:

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier, XGBRegressor
from sklearn.metrics import confusion_matrix, accuracy_score
```

# Importing the dataset:

```python
dataset = pd.read_csv('../data/Breast_cancer_dataset.csv')
```

```python
dataset.head()
```

# Check Missing values

```python
dataset.isna().sum()
```

# Check Duplicates

```python
dataset.duplicated().sum()
```

# Dropping irrelevant features

```python
dataset_selected = dataset.drop(columns=['id', 'Unnamed: 32'], axis=1)
```

```python
dataset_selected.head()
```

```python
dataset_selected.shape
```

# Creating Features

```python
# Creating Feature out of the Dataset:
X = dataset_selected.iloc[:, 1:].values
y = dataset_selected.iloc[:, 0].values
```

```python
print(X)
```

```python
print(y)
```

# Encoding the Variable

As all the feature in this Dataset is a Numerical value, therefore, scaling them using Standard Scaler.

```python

sc = StandardScaler()

X[:, :] = sc.fit_transform(X[:, :])

```

```python
print(X)
```

```python
print(y)
```

# Encoding Target Variable

```python
le = LabelEncoder()

y = le.fit_transform(y)
```

```python
print(y)
```

# Splitting Train Test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state= 1)
```

```python
print(x_train)
```

# Training Models

Logistic Regressor:

```python
logistic_classifier = LogisticRegression(random_state = 0, C=1.0, solver='lbfgs', max_iter=100)

logistic_classifier.fit(x_train, y_train)
```

```python
y_pred_logistic_reg = logistic_classifier.predict(x_test)

print(np.concatenate((y_pred_logistic_reg.reshape(len(y_pred_logistic_reg), 1), y_test.reshape(len(y_test), 1)), axis = 1))
```

```python
cm_logistic_reg = confusion_matrix(y_test, y_pred_logistic_reg)

print(cm_logistic_reg)

accuracy_score(y_test, y_pred_logistic_reg)
```

K-Nearest Neighbors (K-NN)

```python
knn_classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
knn_classifier.fit(x_train, y_train)
```

```python
y_pred_knn_classifier = knn_classifier.predict(x_test)

print(np.concatenate((y_pred_knn_classifier.reshape(len(y_pred_knn_classifier), 1), y_test.reshape(len(y_test), 1)), axis = 1))
```

```python
cm_knn_classifier = confusion_matrix(y_test, y_pred_knn_classifier)

print(cm_knn_classifier)

accuracy_score(y_test, y_pred_knn_classifier)
```

Support Vector Machine (SVM)

```python
svm_classifier = SVC(kernel = 'linear', random_state = 0)

svm_classifier.fit(x_train, y_train)
```

```python
y_pred_svm_classifier = svm_classifier.predict(x_test)

print(np.concatenate((y_pred_svm_classifier.reshape(len(y_pred_svm_classifier), 1), y_test.reshape(len(y_test), 1)), axis = 1))
```

```python
cm_svm_classifier = confusion_matrix(y_test, y_pred_svm_classifier)

print(cm_svm_classifier)

accuracy_score(y_test, y_pred_svm_classifier)
```

Random Forest Classification

```python
random_forest_classifier = RandomForestClassifier(n_estimators= 10, criterion= 'entropy', random_state= 0)

random_forest_classifier.fit(x_train, y_train)
```

```python
y_pred_random_forest = random_forest_classifier.predict(x_test)

print(np.concatenate((y_pred_random_forest.reshape(len(y_pred_random_forest), 1), y_test.reshape(len(y_test), 1)), axis = 1))
```

```python
cm_random_forest = confusion_matrix(y_test, y_pred_random_forest)

print(cm_random_forest)

accuracy_score(y_test, y_pred_random_forest)
```

Decision Tree Classification

```python
decision_tree_classifier = DecisionTreeClassifier(criterion= 'entropy', random_state= 0)

decision_tree_classifier.fit(x_train, y_train)
```

```python
y_pred_decision_tree = decision_tree_classifier.predict(x_test)

print(np.concatenate((y_pred_decision_tree.reshape(len(y_pred_decision_tree), 1), y_test.reshape(len(y_test), 1)), axis = 1))
```

```python
cm_decision_tree = confusion_matrix(y_test, y_pred_decision_tree)

print(cm_decision_tree)

accuracy_score(y_test, y_pred_decision_tree)
```

XGBoost Classifier

```python
XGBoost_classifier = XGBClassifier(use_label_encoder=False, eval_metric='logloss')

XGBoost_classifier.fit(x_train, y_train)
```

```python
y_pred_xgboost = XGBoost_classifier.predict(x_test)

print(np.concatenate((y_pred_xgboost.reshape(len(y_pred_xgboost), 1), y_test.reshape(len(y_test), 1)), axis = 1))
```

```python
cm_xgboost = confusion_matrix(y_test, y_pred_xgboost)

print(cm_xgboost)

accuracy_score(y_test, y_pred_xgboost)
```
